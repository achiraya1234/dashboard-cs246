```python
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ‡∏™‡∏£‡πâ‡∏≤‡∏á DataFrame
data = {
    'Categories': ['‡∏Å‡∏≤‡∏£‡πÄ‡∏î‡∏¥‡∏ô‡∏ó‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏≤‡∏°‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢', '‡∏Å‡∏≤‡∏£‡∏®‡∏∂‡∏Å‡∏©‡∏≤', '‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û', '‡∏™‡∏¥‡πà‡∏á‡πÅ‡∏ß‡∏î‡∏•‡πâ‡∏≠‡∏°'],
    'Rank 1': [a1, b1, c1, d1],
    'Rank 2': [a2, b2, c2, d2],
    'Rank 3': [a3, b3, c3, d3]
}
df = pd.DataFrame(data)

# ‡∏™‡∏£‡πâ‡∏≤‡∏á Heatmap
def create_heatmap(df):
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.set_index('Categories'), cmap='YlGnBu', annot=True, fmt="d", linewidths=.5)
    plt.title('Heatmap of Categories by Rank')
    plt.xlabel('Rank')
    plt.ylabel('Categories')
    st.pyplot()

# ‡πÅ‡∏™‡∏î‡∏á Heatmap ‡πÉ‡∏ô Dashboard
def main():
    st.title('Dashboard with Heatmap')
    st.write("Heatmap of Categories by Rank")

    create_heatmap(df)

if __name__ == '__main__':
    main()
```

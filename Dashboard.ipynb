```python
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# สร้าง DataFrame
data = {
    'Categories': ['การเดินทางและความปลอดภัย', 'การศึกษา', 'สุขภาพ', 'สิ่งแวดล้อม'],
    'Rank 1': [a1, b1, c1, d1],
    'Rank 2': [a2, b2, c2, d2],
    'Rank 3': [a3, b3, c3, d3]
}
df = pd.DataFrame(data)

# สร้าง Heatmap
def create_heatmap(df):
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.set_index('Categories'), cmap='YlGnBu', annot=True, fmt="d", linewidths=.5)
    plt.title('Heatmap of Categories by Rank')
    plt.xlabel('Rank')
    plt.ylabel('Categories')
    st.pyplot()

# แสดง Heatmap ใน Dashboard
def main():
    st.title('Dashboard with Heatmap')
    st.write("Heatmap of Categories by Rank")

    create_heatmap(df)

if __name__ == '__main__':
    main()
```
